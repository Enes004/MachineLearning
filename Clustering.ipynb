```python
import numpy as np
from sklearn.cluster import KMeans

# 1. VERİ (Sensörden gelen noktalar)
# İlk 3 nokta sol altta (Duba 1), son 3 nokta sağ üstte (Duba 2) kümelenmiş gibi düşün
X = np.array([
    [1.0, 2.0], [1.5, 1.8], [1.0, 1.0],  
    [8.0, 8.0], [8.5, 9.0], [9.0, 8.0]   
])

# 2. MODELİ KUR (K=2)
# n_clusters=2 diyerek algoritmayı başlatıyoruz (2 engel arıyoruz)
kmeans = KMeans(n_clusters=2, random_state=42, n_init="auto")

# 3. HESAPLA VE ETİKETLE (Senin yazdığın o iki döngünün çalıştığı satır)
# fit: J maliyetini minimize edip merkezleri bulur. 
# predict: Hangi nokta hangi gruba ait onu söyler.
etiketler = kmeans.fit_predict(X)

# 4. SONUÇLARI ÇEK
merkezler = kmeans.cluster_centers_

print("Noktaların Grupları (c^i):", etiketler)
print("Engellerin Tam Merkezi (Mu_k):\n", merkezler)
```
